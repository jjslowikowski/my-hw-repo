## Part 3. Hand calculating odds ratios

Develop your intuition about expected outcomes by hand calculating odds ratios.

```python
df_keep = ['admit', 'gre', 'gpa']
HC = pd.DataFrame(df[df_keep].join(df.loc[:, 'R1':]))
HC.head()
```

```python
#crosstab prestige 1 admission 
# frequency table cutting prestige and whether or not someone was admitted
ct_1 = pd.crosstab(HC['admit'],HC['R1'])
ct_1
```

#### 3.1 Use the cross tab above to calculate the odds of being admitted to grad school if you attended a #1 ranked college

```python
#overall admits
#looks like this is a quick way to do the rates of admission of the pop...
ct_1 / ct_1.sum()
```

```python
ct_1
```

```python
#odds admit from R1
r1_in = ct_1[1][1]
r1_out = ct_1[1][0]
r1_odds_in = r1_in / r1_out
r1_odds_in
```

#### 3.2 Now calculate the odds of admission if you did not attend a #1 ranked college

```python
not_r1_in = ct_1[0][1]
not_r1_out = ct_1[0][0]
not_r1_odds_in = not_r1_in / not_r1_out
not_r1_odds_in
```

#### 3.3 Calculate the odds ratio

```python
#exposed/unexposed
r1_odds_in / not_r1_odds_in
```

#### 3.4 Write this finding in a sentenance: 

Answer: The odds that an applicant from a #1 ranked college being admitted to grad school is 3.08x higher than an applicant not from a #1 ranked college

#### 3.5 Print the cross tab for prestige_4

```python
ct_4 = pd.crosstab(HC['admit'],HC['R4'])
ct_4
```

#### 3.6 Calculate the OR 

```python
#odds exposed
ct_4_exp = ct_4[1][1] / ct_4[1][0]
#odds unexposed
ct_4_unexp = ct_4[0][1] / ct_4[0][0]
#OR
ct_4_exp / ct_4_unexp
```

## Part 5: Predicted probablities


As a way of evaluating our classifier, we're going to recreate the dataset with every logical combination of input values. This will allow us to see how the predicted probability of admission increases/decreases across different variables. First we're going to generate the combinations using a helper function called cartesian (above).

We're going to use np.linspace to create a range of values for "gre" and "gpa". This creates a range of linearly spaced values from a specified min and maximum value--in our case just the min/max observed values.

```python
def cartesian(arrays, out=None):
    """
    Generate a cartesian product of input arrays.
    Parameters
    ----------
    arrays : list of array-like
    1-D arrays to form the cartesian product of.
    out : ndarray
        Array to place the cartesian product in.
    Returns
    -------
    out : ndarray
        2-D array of shape (M, len(arrays)) containing cartesian products
        formed of input arrays.
    Examples
    --------
    >>> cartesian(([1, 2, 3], [4, 5], [6, 7]))
    array([[1, 4, 6],
           [1, 4, 7],
           [1, 5, 6],
           [1, 5, 7],
           [2, 4, 6],
           [2, 4, 7],
           [2, 5, 6],
           [2, 5, 7],
           [3, 4, 6],
           [3, 4, 7],
           [3, 5, 6],
           [3, 5, 7]])
    """

    arrays = [np.asarray(x) for x in arrays]
    dtype = arrays[0].dtype

    n = np.prod([x.size for x in arrays])
    if out is None:
        out = np.zeros([n, len(arrays)], dtype=dtype)

    m = int(n / arrays[0].size)
    out[:,0] = np.repeat(arrays[0], m)
    if arrays[1:]:
        cartesian(arrays[1:], out=out[0:m,1:])
        for j in np.arange(1, arrays[0].size):
            out[int(j*m):int((j+1))*int(m),1:] = out[0:int(m),1:]
    return out
```

```python
# instead of generating all possible values of GRE and GPA, we're going
# to use an evenly spaced range of 10 values from the min to the max 
gres = np.linspace(Analysis['gre'].min(), Analysis['gre'].max(), 10)
print (gres)
# array([ 220.        ,  284.44444444,  348.88888889,  413.33333333,
#         477.77777778,  542.22222222,  606.66666667,  671.11111111,
#         735.55555556,  800.        ])
gpas = np.linspace(Analysis['gpa'].min(), Analysis['gpa'].max(), 10)
print (gpas)
# array([ 2.26      ,  2.45333333,  2.64666667,  2.84      ,  3.03333333,
#         3.22666667,  3.42      ,  3.61333333,  3.80666667,  4.        ])


# enumerate all possibilities
combos = pd.DataFrame(cartesian([gres, gpas, [1, 2, 3, 4], [1.]]),columns=('gre','gpa','rank','int'))
```
